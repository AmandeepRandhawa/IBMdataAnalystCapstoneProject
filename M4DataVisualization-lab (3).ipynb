## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python


# Open a database connection
conn = sqlite3.connect("m4_survey_data.sqlite")

# Define the query to select the ConvertedComp column
QUERY = """
SELECT ConvertedComp
FROM master
"""

# Execute the query and load the data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Drop any rows with missing or NaN values in ConvertedComp
df = df.dropna(subset=['ConvertedComp'])

# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(df['ConvertedComp'], bins=50, color='blue', edgecolor='black')
plt.title('Histogram of Converted Compensation')
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python

# Open a database connection
conn = sqlite3.connect("m4_survey_data.sqlite")

# Define the query to select the language data from LanguageDesireNextYear table
QUERY = """
SELECT LanguageDesireNextYear, COUNT(*) as Count
FROM LanguageDesireNextYear
GROUP BY LanguageDesireNextYear
ORDER BY Count DESC
"""

# Execute the query and load the data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Close the database connection
conn.close()

# Process and sort the DataFrame to find the rank of Python
df_sorted = df.sort_values(by='Count', ascending=False)

# Find the rank of Python
python_rank = df_sorted[df_sorted['LanguageDesireNextYear'] == 'Python'].index[0] + 1

# Print the rank of Python
print(f"Python is ranked #{python_rank} among the languages respondents wish to learn next year.")
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python

# Open a database connection
conn = sqlite3.connect("m4_survey_data.sqlite")

# Define the query to select WorkWeekHrs, CodeRevHrs, and Age columns
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""

# Execute the query and load the data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Drop any rows with missing or NaN values in WorkWeekHrs, CodeRevHrs, or Age
df = df.dropna(subset=['WorkWeekHrs', 'CodeRevHrs', 'Age'])

# Plot the bubble plot
plt.figure(figsize=(10, 6))
bubble_plot = plt.scatter(x=df['WorkWeekHrs'], y=df['CodeRevHrs'], 
                          s=df['Age'], alpha=0.5, c=df['Age'], cmap='viridis', edgecolor='w', linewidth=0.5)

plt.colorbar(label='Age')
plt.title('Bubble Plot of WorkWeekHrs vs CodeRevHrs (Bubble Size: Age)')
plt.xlabel('Work Hours per Week')
plt.ylabel('Code Review Hours')
plt.grid(True)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Open a database connection
conn = sqlite3.connect("m4_survey_data.sqlite")

# Define the query to select the DatabaseDesireNextYear column
QUERY = """
SELECT DatabaseDesireNextYear
FROM DatabaseDesireNextYear
"""

# Execute the query and load the data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Split the databases listed in each row (assuming the entries are comma-separated)
df = df['DatabaseDesireNextYear'].str.split(';', expand=True).stack().reset_index(drop=True)

# Count the occurrences of each database
db_counts = df.value_counts().reset_index()
db_counts.columns = ['Database', 'Count']

# Get the top 5 databases
top_5_dbs = db_counts.head(5)

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_5_dbs['Count'], labels=top_5_dbs['Database'], autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python

# Define the query to select Age, WorkWeekHrs, and CodeRevHrs columns
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
"""

# Execute the query and load the data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Drop any rows with missing or NaN values in WorkWeekHrs or CodeRevHrs
df = df.dropna(subset=['WorkWeekHrs', 'CodeRevHrs'])

# Compute median values
median_work_week_hrs = df['WorkWeekHrs'].median()
median_code_rev_hrs = df['CodeRevHrs'].median()

# Create data for stacked bar chart
data = {
    'Metric': ['WorkWeekHrs', 'CodeRevHrs'],
    'Median Hours': [median_work_week_hrs, median_code_rev_hrs]
}

df_stack = pd.DataFrame(data)

# Plot the stacked bar chart
plt.figure(figsize=(8, 6))
plt.bar(df_stack['Metric'], df_stack['Median Hours'], color=['skyblue', 'salmon'])
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
plt.xlabel('Metric')
plt.ylabel('Median Hours')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python

# Open a database connection
conn = sqlite3.connect("m4_survey_data.sqlite")

# Define the query to select Age and ConvertedComp columns for ages between 45 and 60
QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
"""

# Execute the query and load the data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Drop any rows with missing or NaN values in ConvertedComp
df = df.dropna(subset=['ConvertedComp'])

# Compute median ConvertedComp for each age
median_comp_by_age = df.groupby('Age')['ConvertedComp'].median().reset_index()

# Plot the median ConvertedComp
plt.figure(figsize=(10, 6))
plt.plot(median_comp_by_age['Age'], median_comp_by_age['ConvertedComp'], marker='o', linestyle='-', color='b')
plt.title('Median Converted Compensation by Age (45 to 60)')
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation')
plt.grid(True)
plt.show()

```

Create a horizontal bar chart using column `MainBranch.`


```python


# Open a database connection
conn = sqlite3.connect("m4_survey_data.sqlite")

# Define the query to select the MainBranch column
QUERY = """
SELECT MainBranch
FROM master
"""

# Execute the query and load the data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Count the occurrences of each value in the MainBranch column
main_branch_counts = df['MainBranch'].value_counts()

# Create a horizontal bar chart
plt.figure(figsize=(10, 6))
main_branch_counts.plot(kind='barh', color='skyblue', edgecolor='black')
plt.title('Distribution of MainBranch')
plt.xlabel('Count')
plt.ylabel('MainBranch')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```
